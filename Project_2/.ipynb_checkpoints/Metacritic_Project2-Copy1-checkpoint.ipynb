```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Last 3 years Metacritic Top 20 Ratings 

base_urlA = 'https://www.metacritic.com/pictures/best-tv-shows-of-2021-at-midyear'
a_base_urlA = np.array([('https://www.metacritic.com/pictures/best-television-shows-of-2020'),('https://www.metacritic.com/pictures/best-new-tv-series-of-2019')])
```

```python
for s in (1,2):
    if(s==1):
        print('******************** YEAR 2020 ********************')
    elif(s==2):
        print('******************** YEAR 2019 ********************')
   
    for x in range(1,21,1):
        
        new_url = a_base_urlA[s-1] + "/" + str(x)
        # print(a_base_url)
        browser.visit(new_url)
        # scrap title, rating, and services
        # Use Beautiful Soup's find() method to navigate and retrieve attributes
        html = browser.html
        # Parse HTML with Beautiful Soup
        soup = BeautifulSoup(html, 'html.parser')

        titles = soup.find_all('span', class_='product_title')
        m_rating = soup.find('span', class_='rank')
        services = soup.find('span', class_='byline inset_left')
        for title in titles:
            print('----------------------------------------------------------------------------------------')
            print(title.text.lstrip())
            print('Rank: ', x)
        

    # Click the 'Next' button on each page
    try:
            browser.links.find_by_partial_text('Next').click()

    except:
            print("Scraping Complete")
            
    print('***************************************************')
    print('Source: ', a_base_urlA[s-1])
    print('***************************************************\n\n\n')
```
