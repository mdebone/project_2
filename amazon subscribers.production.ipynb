```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Prime Video 2019
p19 = Path("Resources/Prime/p_19.csv")
p19_titles = pd.read_csv(p19)
p19_titles.set_index("Title")
```

```python
# Number of New Shows
titles_p19 = len(p19_titles["Title"].unique())
titles_p19
```

```python
# Average IMDb Ranking
p_imdb_19 = p19_titles["IMDb"].mean()
p_imdb_19.round(1)
```

```python
# top 10 Amazon Movies according to IMDB
p_top10_imdb_19 = p19_titles.sort_values("IMDb", ascending=False)
p_top10_imdb_19.set_index("Title").head(10)
```

```python
# Average Rotten Tomatoes Score
p_rot_19 = p19_titles["Rotten Tomatoes"].mean()
p_rot_19.round(1)
```
